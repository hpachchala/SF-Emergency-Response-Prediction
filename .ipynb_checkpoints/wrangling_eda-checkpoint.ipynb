```python
# 1. Imports and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset
df = pd.read_csv("Fire_Department_and_Emergency_Medical_Services_Dispatched_Calls_for_Service_20250530.csv", low_memory=False)
print("Initial shape:", df.shape)
df.head()
```

```python
# Peek at the data
df.info()
df.describe(include='all')
```

```python
# 3. Drop Fully Null Columns
df = df.dropna(axis=1, how='all')

# Check missing values
nulls = df.isna().sum().sort_values(ascending=False)
print(nulls[nulls > 0])
```

```python
# 4. Drop Irrelevant or Sparse Columns
df = df.drop(columns=[
    "Transport DtTm", "Hospital DtTm", "case_location", "RowID",
    "data_as_of", "data_loaded_at"
], errors="ignore")
```

```python
# 5. Clean and Parse Timestamps
df = df.dropna(subset=["Received DtTm", "On Scene DtTm"])
df["Received DtTm"] = pd.to_datetime(df["Received DtTm"], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')
df["On Scene DtTm"] = pd.to_datetime(df["On Scene DtTm"], format="%m/%d/%Y %I:%M:%S %p", errors='coerce')

# Drop rows with invalid timestamps
df = df.dropna(subset=["Received DtTm", "On Scene DtTm"])

# Compute response_time (minutes)
df["response_time"] = (df["On Scene DtTm"] - df["Received DtTm"]).dt.total_seconds() / 60
```

```python
# 6. Remove Outliers 
q_low = df["response_time"].quantile(0.01)
q_high = df["response_time"].quantile(0.99)
df = df[(df["response_time"] > q_low) & (df["response_time"] < q_high)]

print("Response time range:", df["response_time"].min(), "-", df["response_time"].max())
```

```python
# 7. Feature Engineering
df["hour"] = df["Received DtTm"].dt.hour
df["day_of_week"] = df["Received DtTm"].dt.day_name()
df["month"] = df["Received DtTm"].dt.month

# Simplify Call Type (grouping rare categories)
df["Call Type"] = df["Call Type"].fillna("Unknown")
rare_types = df["Call Type"].value_counts()[df["Call Type"].value_counts() < 200].index
df["Call Type (Grouped)"] = df["Call Type"].replace(rare_types, "Other")
```

```python
# 8. Null Summary
print("Null values after cleaning:")
print(df.isna().sum().sort_values(ascending=False))
```

```python
# 9. EDA Visuals

# Distribution of response time
plt.figure(figsize=(8, 4))
sns.histplot(df["response_time"], bins=50, kde=True)
plt.title("Distribution of Emergency Response Time")
plt.xlabel("Response Time (minutes)")
plt.show()

# Boxplot by call type
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x="Call Type (Grouped)", y="response_time")
plt.xticks(rotation=45)
plt.title("Response Time by Call Type")
plt.show()

# Heatmap by hour and weekday
pivot = df.pivot_table(values="response_time", index="day_of_week", columns="hour", aggfunc="mean")
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, cmap="YlOrRd")
plt.title("Avg Response Time by Hour and Day of Week")
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")
plt.show()

# Correlation Matrix
plt.figure(figsize=(20, 20))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()

# Categorical Distribution Check
plt.figure(figsize=(10, 4))
df["Call Type (Grouped)"].value_counts().plot(kind="bar")
plt.title("Call Type Distribution")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 4))
df["Unit Type"].value_counts().plot(kind="bar")
plt.title("Unit Type Distribution")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Time EDA
df_2016 = df[(df["Received DtTm"] >= "2016-01-01") & (df["Received DtTm"] < "2017-01-01")].copy()
df_2016.set_index("Received DtTm", inplace=True)
weekly_avg_2016 = df_2016["response_time"].resample("W").mean()

plt.figure(figsize=(14, 5))
weekly_avg_2016.plot()
plt.title("Weekly Average Emergency Response Time (2016)")
plt.ylabel("Minutes")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 10. Save Cleaned Data
df.to_csv("cleaned_fire_response_data.csv", index=False)
print("Cleaned data saved. Final shape:", df.shape)
```
